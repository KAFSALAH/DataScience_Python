```python
import numpy as np
import pandas as pd
from PIL import Image
import cv2
import matplotlib.pylab as plt
from matplotlib.pylab import figure
```

```python
im1 = Image.open(r"/Users/salahkaf/Desktop/apple1.png")
im2 = Image.open(r"/Users/salahkaf/Desktop/newapple.png")
imagelist = [im1,im2]
im_split=[0,0]
for i in range(0,2):
    im_split[i]= Image.Image.split(imagelist[i])
```

```python
# showing each band for apple
figure(figsize=(8, 6), dpi=80)
plt.subplot(2, 2, 1)
plt
plt.imshow(im1)
plt.title("Original Image")
plt.subplot(2, 2, 2)
plt.imshow(im_split[0][0])
plt.title("Red band")
plt.subplot(2, 2, 3)
plt.imshow(im_split[0][1])
plt.title("Green band")
plt.subplot(2, 2, 4)
plt.imshow(im_split[0][2])
plt.title("Blue Band")
```

```python
# showing each band for apple
figure(figsize=(8, 6), dpi=80)
plt.subplot(2, 2, 1)
plt
plt.imshow(im1)
plt.title("Original Image")
plt.subplot(2, 2, 2)
plt.imshow(im_split[1][0])
plt.title("Red band")
plt.subplot(2, 2, 3)
plt.imshow(im_split[1][1])
plt.title("Green band")
plt.subplot(2, 2, 4)
plt.imshow(im_split[1][2])
plt.title("Blue Band")
```

```python
### Original Apple conversion 
rgb_s0 = im_split[0][0].convert('RGB')
rgb_s1 = im_split[0][1].convert('RGB')
rgb_s2 = im_split[0][2].convert('RGB')
rgb_s0.save('/Users/salahkaf/Desktop/apple2.png')
rgb_s1.save('/Users/salahkaf/Desktop/apple3.png')
rgb_s2.save('/Users/salahkaf/Desktop/apple4.png')
### New apple Conversion
rgb_s0 = im_split[1][0].convert('RGB')
rgb_s1 = im_split[1][1].convert('RGB')
rgb_s2 = im_split[1][2].convert('RGB')
rgb_s0.save('/Users/salahkaf/Desktop/newapple2.png')
rgb_s1.save('/Users/salahkaf/Desktop/newapple3.png')
rgb_s2.save('/Users/salahkaf/Desktop/newapple4.png')
```

```python
img1 = cv2.imread('/Users/salahkaf/Desktop/apple2.png', 0)
img2 = cv2.imread('/Users/salahkaf/Desktop/apple3.png', 0)
img3 = cv2.imread('/Users/salahkaf/Desktop/apple4.png', 0)

figure(figsize=(14, 4), dpi=900)
plt.subplot(1, 3, 1)
plt.hist(img1.ravel(), 256, (5, 230),density=True)
plt.title("Red band")
plt.subplot(1, 3, 2)
plt.hist(img2.ravel(), 256, (5, 230),density=True)
plt.title("Green band")
plt.subplot(1, 3, 3)
plt.hist(img3.ravel(), 256, (5, 230),density=True)
plt.title("Blue Band")
plt.show()
```
