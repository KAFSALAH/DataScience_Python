```python
import numpy as np
import pandas as pd
from PIL import Image
import cv2
from skimage import io
import matplotlib.pylab as plt
from matplotlib.pylab import figure
```

```python
 # opening pictures
im1 = Image.open(r"/Users/salahkaf/Desktop/apple1.png")
im2 = Image.open(r"/Users/salahkaf/Desktop/orange1.png")
im3 = Image.open(r"/Users/salahkaf/Desktop/banana1.png")
im4 = Image.open(r"/Users/salahkaf/Desktop/honeydew1.png")
im5 = Image.open(r"/Users/salahkaf/Desktop/watermelon1.png")
imagelist = [im1,im2,im3,im4,im5]
# split() method
# this will split the image in individual bands
# and return a tuple
im_split=[0,0,0,0,0]
for i in range(0,5):
    im_split[i]= Image.Image.split(imagelist[i])
```

```python
# showing each band for apple
figure(figsize=(8, 6), dpi=80)
plt.subplot(2, 2, 1)
plt
plt.imshow(im1)
plt.title("Original Image")
plt.subplot(2, 2, 2)
plt.imshow(im_split[0][0])
plt.title("Red band")
plt.subplot(2, 2, 3)
plt.imshow(im_split[0][1])
plt.title("Green band")
plt.subplot(2, 2, 4)
plt.imshow(im_split[0][2])
plt.title("Blue Band")
```

```python
img1 = cv2.imread('/Users/salahkaf/Desktop/apple2.png', 0)
img2 = cv2.imread('/Users/salahkaf/Desktop/apple3.png', 0)
img3 = cv2.imread('/Users/salahkaf/Desktop/apple4.png', 0)

figure(figsize=(14, 4), dpi=900)
plt.subplot(1, 3, 1)
axes = plt.gca()
axes.set_ylim([0,10000])
plt.hist(img1.ravel(), 256, (0, 256))
plt.title("Red band")
plt.subplot(1, 3, 2)
axes = plt.gca()
axes.set_ylim([0,10000])
plt.hist(img2.ravel(), 256, (0, 256))
plt.title("Green band")
plt.subplot(1, 3, 3)
axes = plt.gca()
axes.set_ylim([0,10000])
plt.hist(img3.ravel(), 256, (0, 256))
plt.title("Blue Band")
plt.show()
```
